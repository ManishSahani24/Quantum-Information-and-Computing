## Alice and Bob

* Alice and Bob are names given to hypothetical entities or agents in systems, protocols or games that involves exchange of information.

* sender (Alice) transmits a qubit to a receiver (Bob) by making use of a shared entangled quantum state along with two bits of classical communication.

* They are assumed to be in different locations.

$$$$

Let us consider $|\phi^+\rangle$, which is an entangled state of two-qubit.

$$|\phi^+\rangle = \frac{1}{\sqrt{2}}|00\rangle + \frac{1}{\sqrt{2}}|11\rangle$$

Let say this is an e-bit. To say that Alice has a qubit $A$, Bob has a qubit $B$, and together the pair (A,B) is in the state $|\phi^+\rangle$.


* $|00\rangle$ and $|11\rangle$: These represent the basis states of a two-qubit system.

* $|00\rangle$ means Qubit 1 is in state $0$ and Qubit 2 is in state $0$ with probability $1/2$.

* $|11\rangle$ means Qubit 1 is in state $1$ and Qubit 2 is in state $1$ with probability $1/2$.

---

1. If one of two entangled qubits is measured, then the state of the other qubit is somehow instantaneously affected.

2. The state of the two qubits together cannot be described separately.

3. The two qubits somehow maintain a memory of each other.
